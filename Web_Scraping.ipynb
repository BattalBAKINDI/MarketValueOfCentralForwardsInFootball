```python
from bs4 import BeautifulSoup
import selenium
import requests
import re
import numpy as np
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium import webdriver
import time
```

```python
import os
import warnings
warnings.filterwarnings(action='ignore')
chromedriver = "C:/Users/asus/Desktop/chromedriver"
os.environ["webdriver.chrome.driver"] = chromedriver
```

```python
driver = webdriver.Chrome(chromedriver)
```

```python
pages = ["https://www.footballtransfers.com/en/values/players/most-valuable-players/striker-center"]
for page in range(2,41):
  pages.append("https://www.footballtransfers.com/en/values/players/most-valuable-players/striker-center/"+str(page))
```

```python
pages
```

```python
x = 0
result = []
for i in pages:
  URL = i
  driver.get(URL)
  time.sleep(4)
  soup = BeautifulSoup(driver.page_source, "html.parser")

  players_ = soup.find_all("table", {"class":"table table-hover no-cursor table-striped leaguetable mvp-table mb-0"})
  for players in players_:
    player5 = players.find_all("tbody", {"id":"player-table-body"})
    for player in player5:
      player1 = player.find_all("tr")
    for player2 in player1:
      try:
        skill = player2.find("div", {"class":"table-skill__skill"}).text.strip()
      except:
        skill = np.nan
      try:
        pot_skill = player2.find("div", {"class":"table-skill__pot"}).text.strip()
      except:
        pot_skill = np.nan
      try:
        name = player2.find("span", {"class":"d-none"}).text.strip()
      except:
        name = np.nan
      print(name)
      try:
        age = player2.find("td", {"class":"m-hide age"}).text.strip()
      except:
        age = np.nan
      try:
        value = player2.find("span", {"class":"player-tag"}).text.strip()
      except:
        value = np.nan

      link = player2.find("div", {"class":"text"}).a.get("href")        
      time.sleep(1)
      driver.get(link)
      time.sleep(4)
      soup_ = BeautifulSoup(driver.page_source, "html.parser")
        
      time.sleep(1)
        
      info_ = soup_.find_all("div", {"class":"row row-cols-2"})
      for info in info_:
        try:
          foot = info.find_all("strong", {"class":"ttl"})[2].find(text=re.compile("Preferred foot")).findNext().text.strip()
        except:
          foot = np.nan
        try:
          height = info.find_all("strong", {"class":"ttl"})[3].find(text=re.compile("Height")).findNext().text.strip()
        except:
          height = np.nan
        try:
          weight = info.find_all("strong", {"class":"ttl"})[4].find(text=re.compile("Weight")).findNext().text.strip()
        except: weight = np.nan

            
        
      stats_link = soup_.find_all("ul", {"class":"d-flex align-items-center"})
      for stats_ in stats_link:
        stats = stats_.find_all("li")[1].a.get("href")

        time.sleep(1)

        driver.get(stats)
        time.sleep(4)
        soup1 = BeautifulSoup(driver.page_source, "html.parser")

        time.sleep(1)

        stat = soup1.find_all("div", {"class":"tab-pane show active career-statistics-group-1"})
      for stat_ in stat:
        try:
          season22_23_minute = stat_.find_all("tr", {"class":"season-row"})[0].find_all("td")[2].text.strip().replace(".","")
        except:
          season22_23_minute = np.nan
        try:
          season22_23_goal = stat_.find_all("tr", {"class":"season-row"})[0].find_all("td")[3].text.strip()
        except:
          season22_23_goal = np.nan
        try:
          season22_23_assist = stat_.find_all("tr", {"class":"season-row"})[0].find_all("td")[4].text.strip()
        except:
          season22_23_assist = np.nan
        try:
          season21_22_minute = stat_.find_all("tr", {"class":"season-row"})[1].find_all("td")[2].text.strip().replace(".","")
        except:
          season21_22_minute = np.nan
        try:
          season21_22_goal = stat_.find_all("tr", {"class":"season-row"})[1].find_all("td")[3].text.strip()
        except:
          season21_22_goal = np.nan
        try:
          season21_22_assist = stat_.find_all("tr", {"class":"season-row"})[1].find_all("td")[4].text.strip()
        except:
          season21_22_assist = np.nan
      x = x + 1
      print(x)
      result.append([skill,pot_skill,name,age,value,height,weight,foot,season22_23_minute,season22_23_goal,season22_23_assist,season21_22_minute,season21_22_goal,season21_22_assist])
columns = ["skill","pot_skill","name","age","value","height","weight","foot","season22_23_minute","season22_23_goal","season22_23_assist","season21_22_minute","season21_22_goal","season21_22_assist"]
df = pd.DataFrame.from_records(result, columns=columns)
```
