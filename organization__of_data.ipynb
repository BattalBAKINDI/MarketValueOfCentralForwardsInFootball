```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
datacp.drop(columns=["Unnamed: 0"], axis=1, inplace=True)
```

```python
columns = ["season22_23_goal","season22_23_assist","season21_22_goal","season21_22_assist"]
```

* we replace the -  in the data with 0.

```python
for column in columns:
    datacp[column] = datacp[column].replace('-',0)
```

```python
columns_ = ["season22_23_goal","season22_23_assist","season21_22_goal","season21_22_assist"]
```

* I'm converting our numeric data, which is in the object structure, to an int structure.

```python
for column in columns_:
    datacp[column] = datacp[column].astype(int)
```

```python
datacp.info()
```

* I am combining the statistics (minutes, goals and assists) of 2022 and 2023, which are in separate columns, in the same columns.

```python
datacp["minute"] = datacp.season21_22_minute + datacp.season22_23_minute
datacp["goal"] = datacp.season21_22_goal + datacp.season22_23_goal
datacp["asisst"] = datacp.season21_22_assist + datacp.season22_23_assist
datacp.head()
```

```python
datacp.info()
```

```python
datacp.drop(columns=["season22_23_minute","season22_23_goal","season22_23_assist","season21_22_minute","season21_22_goal","season21_22_assist"],axis=1,inplace=True)
```

```python
datacp['value'] = datacp['value'].apply(lambda x: str(x)[1:-1])
datacp['height'] = datacp["height"].apply(lambda x: str(x)[0:-2])
datacp["weight"] = datacp["weight"].apply(lambda x: str(x)[0:-3])
datacp
```
